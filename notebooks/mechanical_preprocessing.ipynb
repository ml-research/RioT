```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import os 

cwd = os.getcwd()


root_path = Path(cwd) / '..' / 'data' / 'mechanical'
meta_path = root_path / 'meta.csv'
folder_prefix = 'experiment_'

train= [1,2, 13,23]
test= [2,3,5,6,9,10] 

meta = pd.read_csv(meta_path, sep=';')
train_rows = meta[meta["Versuchsnummer"].isin(train)]
test_rows = meta[meta["Versuchsnummer"].isin(test)]


def save_npy_from_csv():
    for i, row in meta.iterrows():
        sensors_path = root_path / f'{folder_prefix}{row["Versuchsnummer"]}'/ 'csv' 
        if not sensors_path.exists():
            continue
        count_csv = 0
        for file in os.listdir(sensors_path):
            if file.endswith(".csv"):
                count_csv +=1
                print(file)
                df = pd.read_csv(sensors_path / file)
                a = df.values
                np.save(sensors_path / file[:-4], a)
                print(a.shape)
                
                # break
                # df.plot(x='Zeit', y='Winkelgeschwindigkeit')
                # plt.show()
                # break
            # break
        # delete all csv in folder
        if count_csv != 0:
            os.system(f'rm {sensors_path}/*.csv')
        print(row["Versuchsnummer"])

# for i, row in train_rows.iterrows():
#     sensors_path = root_path / f'{folder_prefix}{row["Versuchsnummer"]}'/ 'csv' 
#     for file in os.listdir(sensors_path):
#         if file.endswith(".csv"):
#             print(file)
#             df = pd.read_csv(sensors_path / file)
#             a = df.values
#             np.save(sensors_path / file[:-4], a)
#             print(a.shape)
            
#             # break
#             # df.plot(x='Zeit', y='Winkelgeschwindigkeit')
#             # plt.show()
#             # break
#         # break
#     # delete all csv in folder
#     os.system(f'rm {sensors_path}/*.csv')
#     print(row["Versuchsnummer"])


save_npy_from_csv()
meta
```
